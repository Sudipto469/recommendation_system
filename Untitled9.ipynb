<h5>Item based similarity</h5>

```python
sim_options = { 'name': 'cosine' ,'user_based':  False}
kf = KFold(n_splits=2)
algo = KNNWithMeans(k =3 , sim_options = sim_options)
best_algo = None
best_rmse = 1000.0
best_pred = None
for trainset, testset in kf.split(data):
    # train and test algorithm.
    algo.fit(trainset)
    predictions = algo.test(testset)
    # Compute and print Root Mean Squared Error
    rmse = accuracy.rmse(predictions, verbose=True)
    if rmse < best_rmse:
        best_rmse= rmse
        best_algo = algo
        best_pred = predictions
```

```python
testSubject  = 1 #user
train_data, test_data = train_test_split(use_df[['UserID', 'MovieID', 'Rating']], test_size = 0.2)
# train_data, test_data = train_test_split(use_df, test_size = 0.2)
trainSet = Dataset.load_from_df(train_data,reader)
trainSet = trainSet.build_full_trainset()
```

```python
simsMatrix = algo.compute_similarities()
```

```python
simsMatrix
```

```python
testUserInnerID = trainSet.to_inner_uid(testSubject)
```

```python
# Get the top K items we rated
testUserRatings = trainSet.ur[testUserInnerID]
kNeighbors = heapq.nlargest(k, testUserRatings, key=lambda t: t[1])
```

```python
kNeighbors
```

```python
# Get similar items to stuff we liked (weighted by rating)
candidates = defaultdict(float)
for itemID, rating in kNeighbors:
    similarityRow = simsMatrix[itemID]
    for innerID, score in enumerate(similarityRow):
        candidates[innerID] += score * (rating / 5.0)
candidates
```

```python
# Build a dictionary of stuff the user has already seen
watched = {}
for itemID, rating in trainSet.ur[testUserInnerID]:
    watched[itemID] = 1
```

```python
genre_cols = ['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
# Get top-rated items from similar users:
pos = 0
for itemID, ratingSum in sorted(candidates.items(), key=itemgetter(1), reverse=True):
    if not itemID in watched:
        movieID = trainSet.to_raw_iid(itemID)
#         print('movieID',type(movieID))
        movie_title = use_df[use_df['MovieID']==movieID]['Title'].values[0]
        print(movie_title,':', ratingSum)
        pos += 1
        if (pos > 10):
            break
```
